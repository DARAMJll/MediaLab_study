```python
import seaborn as sns
```

```python
df = pd.read_csv('data/bike.csv')
df
```

```python
sns.set_theme(rc = {'figure.figsize': (12,6)}, style = 'white')
sns.barplot(data = df, x = 'month', y = 'total', errorbar = None)
plt.show()
```

```python
sns.set_theme(rc = {'figure.figsize': (12,6)}, style = 'white')
sns.barplot(data = df, x = 'month', y = 'total', errorbar = None, hue = 'workingday')
plt.show()
```

```python
sns.set_theme(rc = {'figure.figsize': (12,6)}, style = 'white')
sns.barplot(data = df, x = 'month', y = 'registered', errorbar = None, hue = 'workingday')
plt.show()
```

```python
sns.set_theme(rc = {'figure.figsize': (12,6)}, style = 'white')
sns.lineplot(data = df, x = 'month', y = 'registered', errorbar = None, hue = 'workingday')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize': (10, 5)}, style='white')
sns.barplot(data = df, x = 'quarter', y = 'registered', errorbar = None, hue = 'workingday')
```
